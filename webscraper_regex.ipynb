```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link = 'https://www.flipkart.com/boat-rockerz-650-60-hours-battery-backup-bluetooth-headset/product-reviews/itm482e2b7228cd5?pid=ACCFVWN4Y6N8CU9X&lid=LSTACCFVWN4Y6N8CU9XFYOC8C&marketplace=FLIPKART'
page = requests.get(link)
print(page)
```

```python
soup = bs(page.content, 'html.parser')
print(soup)
```

```python
name = soup.find_all('p', class_ = '_2sc7ZR _2V5EHH')
print(name)
```

```python
cust_name = []

for i in range(0, len(name)) :
    cust_name.append(name[i].get_text())

print(cust_name)
```

```python
cust_name = list(set(cust_name)) # eliminating duplicates
print(cust_name)
```

```python
title = soup.find_all('p', class_ = '_2-N8zT') # use find element and take contents of the class attribute from the highlighted span line
print(title)
```

```python
review_titles = []

for i in range(0, len(title)) :
    review_titles.append(title[i].get_text())

print(review_titles)
```

```python
ratings = soup.find_all('div', class_ = '_3LWZlK _1BLPMq') # use find element and take contents of the class attribute from the highlighted span line
stars=[]
for rate in ratings:
    stars.append(rate.text)
tot=0
avg=0
rates=[]
for star in stars:
    s=star.split(' ')[0]
    rates.append(s)
    tot=tot+float(s)
avg=tot/len(stars)
print("average rating is",avg)
```
